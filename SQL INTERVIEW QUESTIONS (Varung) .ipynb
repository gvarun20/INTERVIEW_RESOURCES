# 1)What is SQL?

A programming language used for interaction with relational database management systems (RDBMS). This includes fetching, updating, inserting, and removing data from tables.

# 2)What types of joins do you know?

1)INNER JOIN:INNER JOIN keyword selects records that have matching values in both tables
    Code SELECT customers.customer_id, orders.order_id
         FROM customers
         INNER JOIN orders ON customers.customer_id = orders.customer_id;
         
         
2)LEFT JOIN:The LEFT JOIN keyword returns all records from the left table and the matched records from the right table
   Code : SELECT customers.customer_id, orders.order_id
          FROM customers
          LEFT JOIN orders ON customers.customer_id = orders.customer_id;
          

3)RIGHT JOIN :The RIGHT JOIN keyword returns all records from the right table and the matched records from the left table.
    Code :SELECT customers.customer_id, orders.order_id
          FROM customers
          RIGHT JOIN orders ON customers.customer_id = orders.customer_id;
          


4)FULL JOIN:The FULL JOIN keyword returns all records when there is a match in either the left or right table. If there is no match, NULL values are returned for columns from the table without a match.
    Code : SELECT customers.customer_id, orders.order_id
           FROM customers
           FULL JOIN orders ON customers.customer_id = orders.customer_id;

# 6)What aggregate functions do you know?

AVG()  returns the average value

SUM()  returns the sum of values

MIN()  returns the minimum value

MAX()  returns the maximum value

COUNT()  returns the number of rows, including those with null values

FIRST()  returns the first value from a column

LAST() returns the last value from a column

# 7)What is the difference between the DROP and TRUNCATE statements?

DROP deletes a table from the database completely, including the table structure and all the associated constraints, relationships with other tables, and access privileges. TRUNCATE deletes all rows from a table without affecting the table structure and constraints. DROP works slower than TRUNCATE. Both are irreversible DDL (Data Definition Language) commands.
